```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Parameters
n_samples = 300
n_features = 2
centers = 4

# Generate  random date
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, cluster_std=1.5, random_state=42)

# Visualize the data
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Dataset for K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt. show()
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):# making 1,2,3,4...10 clusters in each loop and finding wcss
    kmeans=KMeans(n_init=10,n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
    
```

```python
plt.plot(range(1,11),wcss) # plotting wcss against k

# k =4 is an elbow point i.e we should make 4 cluseters
```

```python
kmeans=KMeans(n_init=10,n_clusters=4) # make 4 cluster 4 bcz we know 4 cluster has to be formed using elbow curve
y_kmeans=kmeans.fit_predict(X) # assigns/gives cluster of each data point
```

```python
y_kmeans # assigns cluster to each data poins 
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0 ,1], s = 60, c = 'red', label = 'Cluster1') # cluster 1 give red 
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 60, c = 'blue', label = 'Cluster2' ) # cluster 2 give blue
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 60, c = 'green', label = 'Cluster3' ) 
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 60, c = 'yellow', label = 'Cluster4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans. cluster_centers_[ :, 1], s = 100, c = 'black', label = 'Centroids' )

plt.legend()

plt.show()
```
